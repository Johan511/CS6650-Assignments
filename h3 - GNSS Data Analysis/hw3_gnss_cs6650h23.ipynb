### Task 2: Compute and plot location errors using haversine distance [15 points] <a id='task2'></a>
Calculate the mean location coordinates for all three locations. Assume those three locations as the `groundtruth`. Compute the error for each location fix sample, which is the distance between the location sample and the groundtruth location. You must use haversine distance ([Wiki](https://en.wikipedia.org/wiki/Haversine_formula)). Plot the CDF (cumulative distribution function) of these errors for the three locations. Comment on the variance of the three distributions.


```python
def deg_to_rad(deg):
    rad = math.pi * deg/180
    return rad
```

```python
RADIUS_OF_EARTH = 6.371 * 1_000_000 
def hav(deg):
    return (1-math.cos(deg))/2

def get_harvesine_distance(gps1, gps2):
    phi1, lamda1 = gps1
    phi2, lamda2 = gps2
    valSquareComp1 = hav(phi2-phi1) 
    valSquareComp2 = hav(lamda2-lamda1) * (1-hav(phi1-phi2) -hav(phi1+phi2))
    arcsin = math.asin(math.sqrt(valSquareComp1+ valSquareComp2))
    return 2*RADIUS_OF_EARTH*arcsin
    
    
```

```python
statusHeaders = ["Type","UnixTimeMillis","SignalCount","SignalIndex","ConstellationType","Svid","CarrierFrequencyHz","Cn0DbHz","AzimuthDegrees","ElevationDegrees","UsedInFix","HasAlmanacData","HasEphemerisData","BasebandCn0DbHz"
]
fixHeaders = ["Type","Provider","LatitudeDegrees","LongitudeDegrees","AltitudeMeters",'SpeedMps','AccuracyMeters','BearingDegrees','UnixTimeMillis','SpeedAccuracyMps','BearingAccuracyDegrees','elapsedRealtimeNanos',
              'VerticalAccuracyMeters']
```

```python
fileNames = ["gnss_indoors.txt", "gnss_entrance.txt", "gnss_outdoors.txt"]
gnssData = []
for i in fileNames:
    gnssData.append(pd.read_csv("data/" + i, names = ["Type"] + [i for i in range(1,37)]))
```

```python
def plot_cdf(data,label ,numBins = 10000):
    hist, binEdges = np.histogram(data, numBins)
    pdf = hist/np.sum(hist)
    cdf = np.cumsum(pdf)
    plt.title(label)
    plt.plot(binEdges[1:], cdf)
    plt.xlabel("Distance from mean location")
    plt.ylabel("Cumulative probability")
    plt.show()
    
```

```python
means = []
```

```python
for idx,df in enumerate(gnssData):
    gps = np.array(df[df["Type"] == "Fix"].iloc[:, [2,3]])
    mean = np.mean(gps, axis=0)
    means.append(mean)
    diffs = np.zeros(gps.shape[0])
    for i in range(len(diffs)):
        diffs[i] = get_harvesine_distance(mean, gps[i])
        
    plot_cdf(diffs, fileNames[idx][:-4])
```
